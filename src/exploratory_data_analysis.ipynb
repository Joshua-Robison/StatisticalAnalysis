## Handling Outliers

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d1 = np.loadtxt("../../data/outlier_1d.txt")
d2 = np.loadtxt("../../data/outlier_2d.txt")
d3 = np.loadtxt("../../data/outlier_curve.txt")
print(d1.shape, d2.shape, d3.shape)
```

```python
plt.scatter(d1, np.random.normal(5, 0.2, size=d1.size), label='Dataset 1');
plt.scatter(d2[:,0], d2[:,1], label='Dataset 2');
plt.legend();
```

```python
plt.plot(d3[:,0], d3[:,1], label='Dataset 3');
plt.legend();
```

```python
mean, std = np.mean(d1), np.std(d1)
z_scores = np.abs((d1 - mean) / std)
threshold = 2 # change threshold to change cut-off boundaries
good = z_scores < threshold

print(f'Rejected {(~good).sum()} points')

from scipy.stats import norm

print(f'z-score of 3 corresponds to a probability of {100 * 2 * norm.sf(threshold):0.2f}%')

visual_scatter = np.random.normal(size=d1.size)
plt.scatter(d1[good], visual_scatter[good], label='Good', color='green');
plt.scatter(d1[~good], visual_scatter[~good], label='Bad', color='red');
plt.legend();
```

```python
from scipy.stats import multivariate_normal as mn

mean, cov = np.mean(d2, axis=0), np.cov(d2.T)
threshold = 0.1 / 100 # change threshold to change cut-off boundaries
good = mn(mean, cov).pdf(d2) > threshold

plt.scatter(d2[good,0], d2[good,1], label='Good', color='green');
plt.scatter(d2[~good,0], d2[~good,1], label='Bad', color='red');
plt.legend();
```

```python
xs, ys = d3.T
p = np.polyfit(xs, ys, deg=5)
ps = np.polyval(p, xs)

plt.scatter(xs, ys, label='Data');
plt.plot(xs, ps, 'r--', lw=3, label='Bad Poly Fit');
plt.legend();
```

```python
x, y = xs.copy(), ys.copy()
for i in range(5):
    p = np.polyfit(x, y, deg=5)
    ps = np.polyval(p, x)
    good = y - ps < 3 # only remove positive outliers
    
    x_bad, y_bad = x[~good], y[~good]
    x, y = x[good], y[good]
    
    plt.plot(x, y, '.', label='Used Data');
    plt.plot(x, np.polyval(p, x), 'r--', lw=3, label=f'Poly fit {i}');
    plt.plot(x_bad, y_bad, '.', label='Not Used Data');
    plt.legend();
    plt.show();
    
    if (~good).sum() == 0:
        break
```

```python
from sklearn.neighbors import LocalOutlierFactor

# change hyperparameters to tweak performance
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.01)
good = lof.fit_predict(d2) == 1

plt.scatter(d2[good,0], d2[good,1], label='Good', color='green');
plt.scatter(d2[~good,0], d2[~good,1], label='Bad', color='red');
plt.legend();
```
